```python
import requests
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
url = 'https://shopee.co.id/api/v2/item/get_ratings?filter=0&flag=1&itemid=18216059945&limit=6&offset='
                                                                        #jangan lupa ubah nama file

header = {
    'User-Agent' : 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.134 Mobile Safari/537.36 Edg/103.0.1264.71'
}

comment, rating = [], []
for i in range(1,6):
    if i <= 4:
        for j in range(3):

            r = requests.get(url+(f"{j * 6}&shopid=750804957&type={i}"), headers=header).json()
            print(url+(f"{j * 6}&shopid=750804957&type={i}"))

            for item in r['data']['ratings']:
                comment.append(item['comment'])
                rating.append(item['rating_star'])
    else:
        for j in range(12):

            r = requests.get(url+(f"{j * 6}&shopid=750804957&type={i}"), headers=header).json()
            print(url+(f"{j * 6}&shopid=750804957&type={i}"))

            for item in r['data']['ratings']:
                comment.append(item['comment'])
                rating.append(item['rating_star'])

colom = ['comment', 'rating']
dict_data = dict(zip(colom, (comment, rating)))

data_scrap = pd.DataFrame(dict_data)

file_name = 'jvb3o5w3fb_Produk5.xlsx'
data_scrap.to_excel(file_name)

data_scrap
```
